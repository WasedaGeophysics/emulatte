## FDコインシデントループ

```python
import sys
import numpy as np
import matplotlib.pyplot as plt
sys.path.append('../')
import emulatte.forward as emu

fig = plt.figure(figsize=(16, 8))
ax = fig.add_subplot(211)
ax2 = fig.add_subplot(212)

thicks = np.array([])
dres = np.array([100])
charg = np.array([50])
tconst = np.array([0.8])
fconst = np.array([0.8])
tc = np.array([0, 0, 100])
rc = np.array([100, 0, 200])
freqtime = np.logspace(-3, 7, 10000)
mod = emu.model(thicks)
clp = emu.transmitter('CoincidentLoop', freqtime, current=1, radius=1, turns=10)
mod.add_colecole_params(dres, charg, tconst, fconst)
mod.locate(clp, tc, rc)
ans, f = mod.emulate(hankel_filter='werthmuller201')

resp = ans['h_z'].real
imsp = ans['h_z'].imag

ax.plot(freqtime, resp, "C0-", linewidth=0.8, label='real')
ax.plot(freqtime, -resp, "C0--", linewidth=0.8)
ax.plot(freqtime, imsp, "C1-", linewidth=0.8, label='imaginary')
ax.plot(freqtime, -imsp, "C1--", linewidth=0.8)
ax.grid(which='major', c='#ccc')
ax.grid(which='minor', c='#eee')
ax.tick_params(which='both', direction='in')
ax.set_yscale('log')
ax.set_xscale('log')
ax.set_xlabel('Frequency [Hz]')
ax.set_ylabel('$H_z$ [A/m]')
ax.legend()


thicks = np.array([])
res = np.array([100])
tc = np.array([0, 0, 100])
rc = np.array([100, 0, 200])
freqtime = np.logspace(-3, 7, 10000)

mod = emu.model(thicks)
clp = emu.transmitter('CoincidentLoop', freqtime, current=1, radius=1, turns=10)
mod.add_resistivity(res)
mod.locate(clp, tc, rc)
ans, f = mod.emulate(hankel_filter='werthmuller201')

resp = ans['h_z'].real
imsp = ans['h_z'].imag

ax2.plot(freqtime, resp, "C0-", linewidth=0.8, label='real')
ax2.plot(freqtime, -resp, "C0--", linewidth=0.8)
ax2.plot(freqtime, imsp, "C1-", linewidth=0.8, label='imaginary')
ax2.plot(freqtime, -imsp, "C1--", linewidth=0.8)
ax2.grid(which='major', c='#ccc')
ax2.grid(which='minor', c='#eee')
ax2.tick_params(which='both', direction='in')
ax2.set_yscale('log')
ax2.set_xscale('log')
ax2.set_xlabel('Frequency [Hz]')
ax2.set_ylabel('$H_z$ [A/m]')
ax2.legend()
```

```python
import sys
import numpy as np
import matplotlib.pyplot as plt
sys.path.append('../')
import emulatte.forward as emu


thicks = np.array([20, 5, 2, 5])
res = np.array([100, 80, 10, 80, 100])
tc = np.array([0, 0, 100])
rc = np.array([100, 0, 200])
freqtime = np.logspace(-2, 9, 1000)


mod = emu.model(thicks)
clp = emu.transceiver('CoincidentLoop', freqtime, current=1, radius=1, turns=10)

mod.add_resistivity(res)
mod.locate(clp, tc, rc)
ans, f = mod.emulate(hankel_filter='werthmuller201')

resp = ans['h_z'].real
imsp = ans['h_z'].imag

fig = plt.figure(figsize=(25, 6))
ax = fig.add_subplot(111)

ax.plot(freqtime, resp, "C0-", linewidth=0.5, label='real')
ax.plot(freqtime, -resp, "C0--", linewidth=0.5)
ax.plot(freqtime, imsp, "C1-", linewidth=0.5, label='imaginary')
ax.plot(freqtime, -imsp, "C1--", linewidth=0.5)
ax.grid(which='major', c='#ccc')
ax.grid(which='minor', c='#eee')
ax.tick_params(which='both', direction='in')
ax.set_yscale('log')
ax.set_xscale('log')
ax.set_xlabel('Frequency [Hz]')
ax.set_ylabel('$H_z$ [A/m]')
ax.legend()
```
